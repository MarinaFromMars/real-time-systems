```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 12
w_max = 2700
N = 640
```

```python
def generate(n, w_max, N):
    y = [0 for _ in range(N)]
    x = [i * 0.0001 for i in range(N)]
    w = w_max / n
    for i in range(n):

        w0 = w * (1 + i)
        A = random.random()
        fi = random.random()
        for t in range(N):
            y[t] += A * np.sin(w0 * x[t] + fi)
    return (x, y)
```

```python
x = np.array(generate(n,w_max,N)[0])
y = np.array(generate(n,w_max,N)[1])
result = np.zeros(N,complex)
```

```python
for k in range(N):
    result[k] = np.sum(y * np.exp(-2j*np.pi*k*x))
  
```
